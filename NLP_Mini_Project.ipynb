```python
import nltk
import spacy
from transformers import pipeline
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# Download necessary resources
nltk.download('punkt')
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Initialize Spacy for lemmatization
nlp = spacy.load("en_core_web_sm")

# Preprocessing function to clean the input paragraph
def preprocess_text(text):
    words = word_tokenize(text)
    words = [word for word in words if word.isalnum() and word not in stop_words]
    lemmatized_words = [token.lemma_ for token in nlp(" ".join(words))]
    return " ".join(lemmatized_words)

# Initialize the summarization pipeline with the BART model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarization function
def summarize_text(text):
    # If the input text is too long, you may need to split it into smaller chunks
    # BART can handle approximately 1024 tokens, so chunk text if necessary
    max_chunk_size = 1024

    # Preprocess the input paragraph
    cleaned_text = preprocess_text(text)

    # Perform summarization
    summary = summarizer(cleaned_text, max_length=150, min_length=50, do_sample=False)

    return summary[0]['summary_text']

# Main interaction function
def summarize_paragraph():
    # Take input from the user
    user_input = input("Enter a large paragraph: ")

    # Summarize the input text
    summary = summarize_text(user_input)

    # Print the summary
    print("\nSummary:")
    print(summary)

# Run the summarization function
summarize_paragraph()
```
